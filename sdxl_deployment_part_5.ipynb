## 6. Use the Endpoint for Inference

```python
import boto3
from PIL import Image
from io import BytesIO
import base64
import json
import time

def decode_base64_image(image_string):
    base64_image = base64.b64decode(image_string)
    buffer = BytesIO(base64_image)
    return Image.open(buffer)

client = boto3.client('sagemaker-runtime')

prompt = "A majestic lion, digital art"
negative_prompt = "blurry, bad art, poor quality"

payload = {
    "prompt": prompt,
    "negative_prompt": negative_prompt
}

serialized_payload = json.dumps(payload)

endpoint_name = predictor.endpoint_name

start_time = time.time()

response = client.invoke_endpoint(
    EndpointName=endpoint_name,
    Body=serialized_payload,
    ContentType='application/json'
)

response_payload = json.loads(response['Body'].read().decode("utf-8"))

end_time = time.time()

elapsed_time = end_time - start_time
print(f"Elapsed time: {elapsed_time} seconds")

decoded_image = decode_base64_image(response_payload["image"])
decoded_image.show()
```

## 7. Use the Endpoint for Training

```python
import boto3
import json

runtime = boto3.client('sagemaker-runtime')

endpoint_name = predictor.endpoint_name
content_type = "application/json"

payload = {
    "action": "train",
    "collection_s3_path": "s3://your-bucket/path/to/training/images",
    "prompt": "a photo of sks dog",
    "output_dir_name": "sks_dog_model"
}

response = runtime.invoke_endpoint(
    EndpointName=endpoint_name, 
    ContentType=content_type,
    Body=json.dumps(payload)
)

result = json.loads(response['Body'].read().decode())
print(result)
```
